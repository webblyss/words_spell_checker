```python
import pandas as pd
from textblob import Word
```

```python
df = pd.read_csv('dataset.txt', sep=" ")
df.columns
```

```python
df_list = df.words.values.tolist()
df_list
```

```python
for data in df_list:
    check_spell = Word(data)
    result = check_spell.spellcheck()
    print(result)
```
